### Random tensors:

Why random tensors?

They are important because the way many neural networks learn is that they start with tensors full of random numbers and adjust those random numbers to better represent the data.

`start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

```python
# creating a random tensor using pytorch of size and shape

random_tensor = torch.rand(28,28)
random_tensor
```

```python
random_tensor.ndim, random_tensor.shape
```

```python
another_random_tensor = torch.rand(10,10,10)
another_random_tensor
```

```python
another_random_tensor.ndim, another_random_tensor.shape
```

### Creating a random tensor with simillar shape to an image tensor:

```python
random_image_size_tensor = torch.rand(size =(224,224,3)) # height, width, color channels being red, blue and green
```

```python
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Zeores and Ones:

```python
# create a tensor of all zeroes

zeros = torch.zeros(size =(3,4))
zeros
```

```python
1 * zeros
```

```python
# create a tensor of all ones

ones = torch.ones(size =(3,4))
ones
```

```python
### Creating a range of tensors

torch.arange(0,11)
```

```python
one_to_thousand = torch.arange(start = 1, end = 1000, step = 77)
one_to_thousand
```

```python
one_to_thousand.shape
```

```python
one_to_thousand.ndim
```

```python
one_to_thousand[9]
```

```python
# creating tensors like

ten_zeroes = torch.zeros_like(input = one_to_thousand)
ten_zeroes
```
